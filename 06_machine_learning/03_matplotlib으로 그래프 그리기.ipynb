```python
fig, ax = plt.subplots(figsize =(5, 2.7))
ax.plot(x, x, label = 'linear')
ax.plot(x, x**2, label = 'quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
plt.show()
```

# 선 모양 바꾸기 set_linestyle(모양)

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python

len(data1)
```

```python
np.cumsum(np.arange(1,101))
```

```python
# '-', '--', '-.', ':', ''   => linestyle 
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle="--")
i, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
i.set_linestyle(':')
plt.show()
```

```python
# 반환을 안해주는 경우 뒤에를 비워둘 수 있다?\
```

```python
#많이 사용하지 않는 스타일  
#전처리와 간단한 eda할때는 아래처럼 정성들인 시각화는 하지 않기에 잘 사용안함
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s=100, facecolor='C0', edgecolor='k')
plt.show()
```

# 선 넓이, 선 형태, 마커 모양과 크기

# x축, y축 이름, 눈금, 그래프 이름

```python
mu, sigma =115, 15
x=mu +sigma *np.random.randn(10000)
fig, ax = plt.subplots(figsize =(5,2.7))
n, bins, patches = ax.hist( x, 50, density= True, facecolor='C0', alpha =0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')

ax.text(75, .025, r'$\mu=115, \ \sigma=15$')   #그래프안에 기호 쓰는 법
ax.axis([40, 180, 0, 0.03])    #끝까지 눈금자 범위 지정
# ax.axis('off')       #보통 사진을 불러올때는 꺼준다
ax.grid(True)       #격자 표시
plt.show()
```

# 눈금자 scales

```python
fig, ax = plt.subplots(1,2, figsize =(5,2.7))  #1행 2열 그래프를 만들 것
xdata=np.arange(len(data1))
data = 10**data1
ax[0].plot(xdata, data)
ax[1].plot(xdata, data)
ax[1].set_yscale('log')   #두번째 그래프 스케일 바꿔줘보기
plt.show()
```

# 눈금자 단위 구간 설정하기

```python
fig, ax = plt.subplots(2,1, figsize =(5,2.7), layout ='constrained')   #2행1열 그래프를 만들것
# layout ='constrained' -간격주기
ax[0].plot(xdata, data1)
ax[0].set_title('Automatic ticks')

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0,100,30),['zero','30','sixty','90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title('manual ticks')
plt.show()
```

```python
t= np.arange(0.0,5.0,0.01)
s= np.cos (2* np.pi * t)
fig, (ax1,ax3) =plt.subplots(1,2, figsize=(5,2.7), layout='constrained')
#튜플 형식으로 받기
l1, = ax1.plot(t,s)
ax2 = ax1.twinx()
l2, =ax2.plot(t, range(len(t)),'C1')
ax2.legend([l1,l2],['sine(left)','straight (right)'])

ax3.plot(t, s)
ax3.set_xlabel('Angle [red]')
ax4=ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax4.set_xlabel('Angle [˚]')
```
