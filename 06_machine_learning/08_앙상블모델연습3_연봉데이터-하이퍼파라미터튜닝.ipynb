모든 변환과정을 똑같이 맞춰줘야한다
데이터 금액 등 편차 크면 스케일링 해줘야하지? 그런 작업들 모두
모델만들때 사용한 모든 작업 ㄱ대로 옮겨가야해.
새로운 서비스 받았을때 규격이 안맞아서 엉뚱하게 될것

모델 훈련도 중요하지만 훈련해서 만들어지는 모델 테스에 넣을 때 똑같은 상태로 만들어질수 있는지 꼭 생각하고 작업해라

```python

def ages(x):
    if 10 <= x < 20:
        return "10s"
    elif 20 <= x < 30:
        return "20s"
    elif 30 <= x < 40:
        return "30s"
    elif 40 <= x < 50:
        return "40s"
    elif 50 <= x < 60:
        return "50s"
    elif 60 <= x < 70:
        return "60s"
    elif 70 <= x < 80:
        return "70s"
    elif 80 <= x < 90:
        return "80s"
    elif 90 <= x < 100:
        return "90s"
```

```python
data['age'] = data['age'].apply(ages)
```
