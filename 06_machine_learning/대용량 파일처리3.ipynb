```python
import pandas as pd
data= pd.read_csv("C:/fintech_service/06_machine_learning/data/final_fliter_data2.csv")
data
```

```python
learning_place = pd.read_csv("C:/fintech_service/06_machine_learning/data/학원교습소정보_2024년08월31일기준.csv", encoding='cp949')
learning_place
```

```python
learn_place=learning_place[['시도교육청명', '행정구역명']]
learn_place
```

```python
# '시도교육청' 컬럼에서 '전라북도교육청'을 '전북특별자치도교육청'으로 변경
learn_place['시도교육청명'] = learn_place['시도교육청명'].replace('전라북도교육청', '전북특별자치도교육청')

# 변경된 값 확인
learn_place['시도교육청명'].unique()
```

```python
# '시도교육청' 값의 오른쪽에서 세 글자 제거
learn_place['시도교육청명'] = learn_place['시도교육청명'].str[:-3]
learn_place
# '시도교육청'과 '행정구역명' 데이터를 합쳐서 '시군구' 컬럼 생성
learn_place['시군구'] = learn_place['시도교육청명'] + ' ' + learn_place['행정구역명']
learn_place
```

```python
# '시군구' 컬럼의 값별로 개수 세기
count_per_sigungu = learn_place['시군구'].value_counts().reset_index()

# 데이터프레임의 컬럼명 변경
count_per_sigungu.columns = ['시군구', '학원수']

count_per_sigungu
```

```python
count_per_sigungu['시군구'].unique()
```

```python
# count_per_sigungu.to_csv("C:/fintech_service/06_machine_learning/data/learn_place_count_data.csv", index=False)
```

```python
count_per_sigungu['학원수'].unique()
```

```python
# merged_final에 '분류' 데이터를 기준으로 count_per_sigungu의 '학원수' 데이터를 left join
merged_final = data.merge(count_per_sigungu[['시군구', '학원수']],
                                  left_on='분류', 
                                  right_on='시군구', 
                                  how='left')

merged_final
```

```python
# '시군구_y' 컬럼 삭제
merged_final = merged_final.drop(columns=['시군구_y'])
merged_final
```

```python
merged_final = merged_final.drop(columns=['분류'])
merged_final
```

```python
# '시군구_x' 컬럼명을 '시군구'로 변경
merged_final = merged_final.rename(columns={'시군구_x': '시군구'})

merged_final
```

```python
# 필요한 경우 병합된 데이터를 저장
merged_final.to_csv("C:/fintech_service/06_machine_learning/data/final_dataset.csv", index=False)
```
