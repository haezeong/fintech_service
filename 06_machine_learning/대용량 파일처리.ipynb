```python
import pandas as pd
from sqlalchemy import create_engine

# 데이터베이스 연결 설정
engine = create_engine('mysql+pymysql://root:1234@127.0.0.1:3306/korea_exchange_rate')

# 환율 데이터를 쿼리로 가져옴
query = """
SELECT date, `현찰_살때_환율` AS usd_rate
FROM exchange_rate
WHERE 통화 = '미국 USD'
AND date BETWEEN '1999-01-01' AND '2024-12-31'
"""
usd_data = pd.read_sql(query, engine)

# 날짜 형식 변환
usd_data['date'] = pd.to_datetime(usd_data['date'])
usd_data
```

```python
import pandas as pd

# data 파일 읽기
data = pd.read_csv("C:/fintech_service/06_machine_learning/data/combined_apartment_data2.csv")

# usd_data 파일 읽기 (USD 데이터가 포함된 파일이라고 가정)
# 예시: usd_data = pd.read_csv('path_to_usd_data.csv')

# 'date' 컬럼을 datetime 형식으로 변환
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d', errors='coerce')
usd_data['date'] = pd.to_datetime(usd_data['date'], format='%Y-%m-%d', errors='coerce')

# usd_data에서 'date' 기준으로 중복된 값 제거 (중복된 경우 첫 번째 값만 남김)
usd_data = usd_data.drop_duplicates(subset='date')

# 'date' 컬럼을 기준으로 data와 usd_data 병합 (left join)
merged_data = pd.merge(data, usd_data[['date', 'usd_rate']], on='date', how='left')

# 결측치(NaN)가 있을 경우 이전 날짜 데이터를 채워넣기 (ffill)
merged_data.fillna(method='ffill', inplace=True)

# 결과 확인
merged_data.shape
```

```python
# 필요한 경우 병합된 데이터를 저장
merged_data.to_csv("C:/fintech_service/06_machine_learning/data/merged_apartment_usd_data.csv", index=False)
```
