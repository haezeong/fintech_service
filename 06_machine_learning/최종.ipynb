```python
import pandas as pd
data= pd.read_csv("C:/fintech_service/06_machine_learning/data/precision_data.csv", encoding='UTF8')
data
```

```python
# 원핫 인코딩 적용
data = pd.get_dummies(data, columns=['시군구'], drop_first=True)
```

```python
# 100회 이하로 등장하는 시군구를 '기타'로 그룹화
threshold = 100
freq_counts = data['시군구'].value_counts()

# 빈도수가 threshold 이하인 값들을 '기타'로 변환
data['시군구'] = data['시군구'].apply(lambda x: x if freq_counts[x] > threshold else '기타')

# 원핫 인코딩 적용
data = pd.get_dummies(data, columns=['시군구'], drop_first=True)

# 데이터 확인
print(data.info())
```
