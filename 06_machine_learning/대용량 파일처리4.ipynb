```python
# '분류' 컬럼에서 포항의 각 구를 '경상북도 포항시'로 변경
final['분류'] = final['분류'].replace({
    '경상북도 포항남구': '경상북도 포항시',
    '경상북도 포항북구': '경상북도 포항시'
})
```

```python
# '분류' 컬럼 데이터를 처리하는 함수 정의
def process_region(region):
    parts = region.split()  # ' '을 기준으로 나누기
    if len(parts) > 1 and len(parts[1]) >= 5:  # 두 번째 값의 길이가 5 이상일 경우
        return parts[0] + ' ' + parts[1][:2] + '시'  # 왼쪽에서 두 글자만 가져와서 '시' 붙이기
    return region

# '분류' 컬럼에 대해 함수 적용
final['분류'] = final['분류'].apply(process_region)
```

```python
# merged_final에 '분류' 데이터를 기준으로 count_per_sigungu의 '학원수' 데이터를 left join
merged_final =final.merge(learning[['시군구', '학원수']],
                                  left_on='분류', 
                                  right_on='시군구', 
                                  how='left')

merged_final
```
