```python
import pandas as pd
data= pd.read_csv("C:/fintech_service/06_machine_learning/data/final_merge_data.csv")
data
```

```python
# usd_rate 컬럼명을 달러환율로 변경
data.rename(columns={'usd_rate': '달러환율'}, inplace=True)
data
```

```python
base_rate= pd.read_csv("C:/fintech_service/06_machine_learning/data/flitered_base_rate_data.csv")
base_rate
```

```python
# date 컬럼을 datetime 형식으로 변환 (만약 아직 변환하지 않았다면)
data['date'] = pd.to_datetime(data['date'])
base_rate['date'] = pd.to_datetime(base_rate['date'])

# base_rate와 merged_data를 병합 (가장 가까운 과거의 기준금리 사용)
merged_final = pd.merge_asof(data.sort_values('date'),
                             base_rate.sort_values('date'),
                             on='date', 
                             direction='backward')

# 결과 확인
merged_final
```

```python
# 필요한 경우 병합된 데이터를 저장
merged_final.to_csv("C:/fintech_service/06_machine_learning/data/final_fliter_data.csv", index=False)
```

```python
# '시도교육청' 컬럼에서 '전라북도교육청'을 '전북특별자치도교육청'으로 변경
learn_place['시도교육청명'] = learn_place['시도교육청명'].replace('전라북도교육청', '전북특별자치도교육청')

# 변경된 값 확인
learn_place['시도교육청명'].unique()
```

```python
# '시도교육청' 값의 오른쪽에서 세 글자 제거
learn_place['시도교육청명'] = learn_place['시도교육청명'].str[:-3]
learn_place
```

```python
# '시도교육청'과 '행정구역명' 데이터를 합쳐서 '시군구' 컬럼 생성
learn_place['시군구'] = learn_place['시도교육청명'] + ' ' + learn_place['행정구역명']
learn_place
```

```python
# '시군구' 컬럼의 값별로 개수 세기
count_per_sigungu = learn_place['시군구'].value_counts()
count_per_sigungu
```

```python
# '시군구' 컬럼의 값별로 개수 세기
count_per_sigungu = learn_place['시군구'].value_counts().reset_index()

# 데이터프레임의 컬럼명 변경
count_per_sigungu.columns = ['시군구', '학원수']

count_per_sigungu
```

```python
merged_final
```

```python
# '시군구' 칼럼을 공백(' ') 기준으로 나누고, 왼쪽에서 두 개의 요소만 선택
merged_final['분류'] = merged_final['시군구'].apply(lambda x: ' '.join(x.split()[:2]))
merged_final
```

```python
merged_final['분류'].unique()
```

```python
# '분류' 칼럼을 ' ' 기준으로 나눈 후 첫 번째 값이 '전라북도'인 경우 '전북특별자치도'로 변경
merged_final['분류'] = merged_final['분류'].apply(lambda x: x.replace('전라북도', '전북특별자치도') if x.startswith('전라북도') else x)
merged_final
```
