```python
import requests
import os
import google.generativeai as genai
import gradio as gr
from dotenv import load_dotenv
from transformers import pipeline

# API 키 불러오기
load_dotenv(".chatenv")
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
NAVER_CLIENT_ID = os.getenv('NAVER_CLIENT_ID')
NAVER_CLIENT_SECRET = os.getenv('NAVER_CLIENT_SECRET')

# Google Gemini API 설정
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-1.5-flash')

# 요약 모델 설정
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# 네이버 뉴스 검색 함수
def search_naver_news(query, display=5):
    url = "https://openapi.naver.com/v1/search/news.json"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET,
    }
    params = {
        "query": query,
        "display": display,
        "sort": "date"
    }
    response = requests.get(url, headers=headers, params=params)

    # 응답 상태 코드 확인
    if response.status_code == 200:
        return response.json()['items']
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# 뉴스 요약 함수
def summarize_news(content):
    summary = summarizer(content, max_length=60, min_length=20, do_sample=False)
    return summary[0]['summary_text']

# 질문에 대한 답변 생성 함수
def generate_answer(question):
    response = model.generate_content(question)
    return response.text

# 네이버 뉴스 검색 후 요약 함수 (Markdown 형식으로 보기 좋게 출력)
def search_and_summarize(query):
    news_items = search_naver_news(query)
    
    if not news_items:
        return "No news found."

    news_list = []
    for item in news_items:
        title = item['title']
        link = item['link']
        description = item['description']
        summary = summarize_news(description)
        news_list.append(f"**Title:** {title}\n\n**Summary:** {summary}\n\n[Read more]({link})")
    
    return "\n\n---\n\n".join(news_list)

# Gradio 인터페이스 구성
with gr.Blocks() as app:
    gr.Markdown("# AI Chatbot and News Summarizer")

    # 질문과 답변 챗봇
    with gr.Row():
        question = gr.Textbox(label="Ask a question")
        answer = gr.Textbox(label="Answer")
    ask_btn = gr.Button("Ask")
    ask_btn.click(fn=generate_answer, inputs=question, outputs=answer)

    # 뉴스 검색과 요약
    with gr.Row():
        news_query = gr.Textbox(label="Search Naver News")
        news_output = gr.Markdown(label="News Summary")  # Markdown 형식으로 결과 출력
    search_btn = gr.Button("Search and Summarize News")
    search_btn.click(fn=search_and_summarize, inputs=news_query, outputs=news_output)

# Gradio 앱 실행
app.launch(inline=False, share=True)
```
