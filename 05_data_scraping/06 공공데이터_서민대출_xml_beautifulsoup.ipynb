```python
result = []
page = 1

while True:
    url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    payload = dict(serviceKey=os.getenv('servicekey'), pageNo=page, numOfRows=1000)
    r = requests.get(url, params=payload)
    print(r.status_code)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    total_pages = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    row = {}
    # XML의 각 요소를 반복하며 데이터 추출
    for elem in soup.find_all('item'):
        for tag in elem.find_all():
            row.setdefault(tag.name, []).append(tag.text)
    result.append(row)
    if page < total_pages:
            page += 1
    else:
        break




# 데이터를 pandas DataFrame으로 변환
final_result = pd.DataFrame()
for result_item in result:
    temp = pd.DataFrame(result_item)
    final_result = pd.concat([final_result, temp])
```
