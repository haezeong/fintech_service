```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
import time
from bs4 import BeautifulSoup as bs
load_dotenv()
```

```python
url = f"http://openapi.seoul.go.kr:8088/{os.getenv('seoul_s_key')}/xml/dreamMoney/1/5/"

r = requests.get(url)
soup = bs(r.text, 'xml')
print(r.status_code)
print(r.text)
```

```python
for item in soup.find_all('row'):
    print(item.name)
```

```python
for item in soup.find_all('row'):
    for tag in item:
        print(tag.name)
```

```python
keys={}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue
        else:
            keys.setdefault(tag.name,"")
for key in keys:
    print(soup.find_all(key))
```

```python
keys={}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue
        else:
            keys.setdefault(tag.name,"")
result = {}
for key in keys:
    for item in soup.find_all(key):
        result.setdefault(item.name, []).append(item.string)

result_df = pd.DataFrame(result)
result_df
```

```python
result_dfs = []
page = 1

while True:
    url = f"http://openapi.seoul.go.kr:8088/{os.getenv('seoul_s_key')}/xml/dreamMoney/1/5/"
    payload = dict(serviceKey=os.getenv('seoul_s_key'),TYPE=page, SERVICE='dreamMoney',START_INDEX=1, END_INDEX=)
    r = requests.get(url, params=payload)
    print(r.status_code)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    total_pages = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    keys = {}
    for tags in soup.find_all()[10:]:
        if tags.name == 'item':
            continue
        else:
            keys.setdefault(f"{tags.name}", "")
    result2 = {}
    for key in keys:
        for item in soup.find_all(key):
            result2.setdefault(key, []).append(item.string)
    result_dfs.append(pd.DataFrame(result2))

    if page < total_pages:
            page += 1
    else:
        break

final_df = pd.DataFrame()           
for df in result_dfs:
    final_df = pd.concat([final_df, df])
final_df
```
