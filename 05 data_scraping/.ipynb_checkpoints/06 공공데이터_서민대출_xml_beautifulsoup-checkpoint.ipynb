```python
tag_dict = {}

# 모든 태그 반복
for tag in soup.find_all('item')[0]:
    tag_name = tag.name
    tag_text = tag.text.strip()

    # 태그 이름이 이미 딕셔너리에 있는 경우
    if tag_name in tag_dict:
        tag_dict[tag_name].append(tag_text)
    else:
        tag_dict[tag_name] = [tag_text]

# 딕셔너리 출력
print(tag_dict)
```

# 해결법 : beautifulSoup 으로 html, xml 파싱하기

```python
#request로 가져온것은 다 문자열이기에 xml로 변환한다음에 가져오거나, html태그에서 가져오는 법
```

```python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')   #클래스 인스턴스에 넣어주기

print(soup.prettify())
```

```python
keys={}

for tags in soup.find_all():#중복해서 들어간다
    print(tags.name) #앞에 10개 없애야함
    keys.setdefault(f"{tags.name}", "")
print(len(keys))
```

```python
keys={}

for tags in soup.find_all()[10:]:#중복해서 들어간다
    
    keys.setdefault(f"{tags.name}", "")

print(len(keys))
```
