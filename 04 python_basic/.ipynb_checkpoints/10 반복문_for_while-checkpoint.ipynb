# 반복문 for, while
* 반복문은 동일한 작업을 여러 번 반복해야 할 경우 사용하는 명령
* for, while 두 종류의 반복문이 있음.
* for는 반복 횟수가 정해져 있는 경우에 사용.
* while은 반복 횟수가 정해지지 않은 경우에 사용.

## for 반복문: 반복할 횟수가 정해져 있는 경우
* for 변수 in 반복할 객체, 자료:<br>
    반복할 코드

```python
list(range(1,11))
```

```python
# for문을 이용해서 1-10 숫자 출력
for num in range(1,11):
    print(num)
```

```python
# for문을 이용해서 1-10 숫자 출력
for num in range(1,11):
    print(num, end=" ")
```

enumerate(): index를 만들어 주는 함수

```python
#인덱스를 자료와함께 같이 뽑고 싶을 때 ->enumerate()

for index, char in enumerate("파이썬은 문자열을 반복문에 넣어 출력 가능"):
    print(index, char)
    
#튜플로 묶여서 인덱스를 같이 찍어준다.
```

## 반복 횟수가 정해지지 않은 경우: 무한반복

```python
#조건으로 몇번만 반복하라고 하면되는데, 데이터가 몇개 있는지 모를대, 조건 모를 때,문한 반복을 사용한다
#break로 빠져나와서 중단시키는 방법
#내가 얼마까지 반복시킬지를 모를때 무한반복을 사용한다
#밑에 조건을 달아준다
#ex)밑에 값이 안오면 중단시켜라
#일단 무한반복으로 코드 짜두고 더이상 값이 들어오지 않을때 멈춰라.
#멈출 수 있는 조건을 주어야함
```

무한 반복을 하다가 숫자가 100이 되는 시점까지 숫자를 더해서 합을 구하세요.

```python
i=1
hap=0
while True:
    if i <=100:
        hap+=i
       
    else:
        break
    i+=1
    
print(hap)
```

* True가 되는 경우: 1, 정수, 실수, 문자열, True
* False가 되는 경우: 0,  0.0,  '',  "", [], (), {}

```python
while []:
    print("무한반복")
```

### break 반복문을 정지시키는 명령

```python
for i in range(1,101):
    if i>50:
        break
    else:
        print(i, end=" ")
```

```python
i=1
while i <=100:
    if i>50:
        break
    else:
        print(i, end=" ")
    i += 1
```

```python
#break는 특정 조건을 주고 그 것을 만족하면 멈춰라
```

### continue: 특정 조건에서 코드 실행 건너뛰기

```python
#1~100 까지 수 중에서 짝수면 더하고 홀수면 건너뛰기
hap=0
for i in range(1,101):
    if i % 2 ==1:
        continue
    else:
        hap += i
        print(hap)
        
```

```python
#1~100 까지 수 중에서 짝수면 더하고 홀수면 건너뛰기
hap=0
for i in range(1,101):
    if i % 2 ==1:
        continue
    
    hap += i
    print("i: ",i, "hap:", hap)
        
```

```python
#1~100 까지 수 중에서 짝수면 더하고 홀수면 건너뛰기 -while
hap=0
i=1
while i <=100:
    if i%2==1:
        i += 1
        continue
    else:
        hap += i
        i += 1
        
print("hap:", hap)
```

```python
#1~100 까지 수 중에서 짝수면 더하고 홀수면 건너뛰기 -while
hap=0
i=1
while i <=100:
    if i%2==1:
        i += 1
        
    print(i,"hap:", hap)
```

### 중첩반복문
* 

```python
#시계 만들기
import time
for si in range(0,24): #시침
    for bun in range(0,60): #분침
        for cho in range (0,60): #초침
            time.sleep(1)
            print(si, "시", bun, "분", cho, "초", end ="\r")
```

```python
# 반복문은 최대한 중첩되지 않도록 코딩하기
# 안에 것이 다 돌고 바깥으로 나오기에 계산량이 기하급수적으로 증가하게 됨
# 안쪽부터 시작해서 밖으로 - 마치 시계같음
#계산량이 많아지므로 depth를 깊게 만들지 말자
```
